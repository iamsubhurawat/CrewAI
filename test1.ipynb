````python
from crewai                               import Agent, Task, Crew, Process
from langchain_groq                       import ChatGroq
````

---

# Agents

---

````python
extension_developer = Agent(
role='Senior Developer',
goal='You have to write a code to develop a wordpress extension to display tablular data using rest api ',
backstory="""You work at a leading tech company Webkul.
Your expertise lies in developing wordpress extensions by writing whole code of the files required to create a wordpress extension.""",
verbose=True,
allow_delegation=False,
llm=ChatGroq(model="llama3-70b-8192",temperature=0),
)

php_developer = Agent(
    role='Senior PHP Developer',
    goal='You have to write a php code to develop a wordpress extension given to you by extension_developer agent. Do not add any additional information just the php code only.',
    backstory="""You work at a leading tech company Webkul.
    Your expertise lies in creating only the php portion of a wordpress extension code by writing only php code to create a wordpress extension given to you by extension_developer agent.""",
    verbose=True,
    allow_delegation=False,
    llm=ChatGroq(model="llama3-70b-8192",temperature=0),
)

php_formatter = Agent(
    role='Senior PHP Formatter',
    goal='You have to format a php code to develop a wordpress extension given to you by php_developer agent. Remove "```" from start and end of the code.',
    backstory="""You work at a leading tech company Webkul.
    Your expertise lies in formatting the php code of a wordpress extension given to you by php_developer agent.""",
    verbose=True,
    allow_delegation=False,
    llm=ChatGroq(model="llama3-70b-8192",temperature=0),
)

css_developer = Agent(
    role='Senior CSS Developer',
    goal='You have to write a css code to develop a wordpress extension given to you by extension_developer agent. Do not add any additional information just the css code only.',
    backstory="""You work at a leading tech company Webkul.
    Your expertise lies in creating only the css portion of a wordpress extension code by writing only css code to create a wordpress extension given to you by extension_developer agent.""",
    verbose=True,
    allow_delegation=False,
    llm=ChatGroq(model="llama3-70b-8192",temperature=0),
)

css_formatter = Agent(
    role='Senior CSS Formatter',
    goal='You have to format a css code to develop a wordpress extension given to you by css_developer agent. Remove "```" from start and end of the code.',
    backstory="""You work at a leading tech company Webkul.
    Your expertise lies in formatting the css code of a wordpress extension given to you by css_developer agent.""",
    verbose=True,
    allow_delegation=False,
    llm=ChatGroq(model="llama3-70b-8192",temperature=0),
)

js_developer = Agent(
    role='Senior JS Developer',
    goal='You have to write a js code to develop a wordpress extension given to you by extension_developer agent. Do not add any additional information just the js code only.',
    backstory="""You work at a leading tech company Webkul.
    Your expertise lies in creating only the js portion of a wordpress extension code by writing only js code to create a wordpress extension given to you by extension_developer agent.""",
    verbose=True,
    allow_delegation=False,
    llm=ChatGroq(model="llama3-70b-8192",temperature=0),
)

js_formatter = Agent(
    role='Senior jS Formatter',
    goal='You have to format a js code to develop a wordpress extension given to you by js_developer agent. Remove "```" from start and end of the code.',
    backstory="""You work at a leading tech company Webkul.
    Your expertise lies in formatting the js code of a wordpress extension given to you by js_developer agent.""",
    verbose=True,
    allow_delegation=False,
    llm=ChatGroq(model="llama3-70b-8192",temperature=0),
)
````

---

# Tasks

---

````python
code_generation = Task(
    description="Generating full code to create a wordpress extension to display tablular data using rest api ",
    expected_output="Well crafted codes separately inside proper file with proper name and file extension",
    agent=extension_developer,
    output_file=f"draft.txt",
)

php_code_generation = Task(
    description="Generating only the php code for creating a wordpress extension, given to you by extension_developer agent. Only write php code and do not add any additional information.",
    expected_output="Well crafted php code separately inside proper file with proper name and file extension. Do not add any additional information.",
    agent=php_developer,
    output_file=f"draft.php",
    context=[code_generation]
)

php_code_formatting = Task(
    description="Formatting the php code for creating a wordpress extension, given to you by php_developer agent.",
    expected_output="Well formatted php code only without any irrelevant text. Remove '```' from start and end of the code.",
    agent=php_formatter,
    output_file=f"draft.php",
    context=[php_code_generation]
)

css_code_generation = Task(
    description="Generating only the css code for creating a wordpress extension, given to you by extension_developer agent. Only write css code and do not add any additional information.",
    expected_output="Well crafted css code separately inside proper file with proper name and file extension. Do not add any additional information.",
    agent=css_developer,
    output_file=f"draft.css",
    context=[code_generation]
)

css_code_formatting = Task(
    description="Formatting the css code for creating a wordpress extension, given to you by css_developer agent.",
    expected_output="Well formatted css code only without any irrelevant text. Remove '```' from start and end of the code.",
    agent=css_formatter,
    output_file=f"draft.css",
    context=[css_code_generation]
)

js_code_generation = Task(
    description="Generating only the js code for creating a wordpress extension, given to you by extension_developer agent. Only write js code and do not add any additional information.",
    expected_output="Well crafted js code separately inside proper file with proper name and file extension. Do not add any additional information.",
    agent=js_developer,
    output_file=f"draft.js",
    context=[code_generation]
)

js_code_formatting = Task(
    description="Formatting the js code for creating a wordpress extension, given to you by js_developer agent.",
    expected_output="Well formatted js code only without any irrelevant text. Remove '```' from start and end of the code.",
    agent=js_formatter,
    output_file=f"draft.js",
    context=[js_code_generation]
)
````

---

# Crew

---

````python
crew = Crew(
agents=[extension_developer,php_developer,css_developer,js_developer,php_formatter,css_formatter,js_formatter],
tasks=[code_generation,php_code_generation,css_code_generation,js_code_generation,php_code_formatting,css_code_formatting,js_code_formatting],
process=Process.sequential,
verbose=1,
full_output=True
)
````

````python
result = crew.kickoff()
````
